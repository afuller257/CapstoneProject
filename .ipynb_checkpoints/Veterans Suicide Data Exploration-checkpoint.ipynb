```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas dataframe
df = pd.read_csv("combined_cleaned.csv")

# Print the first 5 rows of the dataframe
print(df.head())

# Get the summary statistics of the numerical columns
print(df.describe())

# Check for missing values in the dataframe
print(df.isnull().sum())

# Plot histograms of multiple numerical columns
df[['2005_vet_suicides','2006_vet_suicides','2007_vet_suicides','2008_vet_suicides','2009_vet_suicides','2010_vet_suicides','2011_vet_suicides']].hist()

# Show the histograms
plt.show()

# Concatenate the columns
combined = pd.concat([df['2005_vet_suicides'], df['2006_vet_suicides'], df['2007_vet_suicides'], df['2008_vet_suicides'], df['2009_vet_suicides'], df['2010_vet_suicides'], df['2011_vet_suicides']])

# Plot the histogram
plt.hist(combined, bins=10)
plt.xlabel('Number of Suicides')
plt.ylabel('Frequency')
plt.title('Histogram of Suicides from 2005 to 2011')
plt.show()


# Group the data by state and sum the suicide rates for each year
df_by_state = df.groupby(['state']).sum()

# Plot a histogram for each state
for state in df_by_state.index:
    df_by_state.loc[state].hist()
    plt.title(state)
    plt.show()

# Plot a scatter plot of two numerical columns
df.plot.scatter(x='2005_vet_suicides', y='2011_vet_suicides')
```
