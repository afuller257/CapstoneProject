```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('combined_cleaned.csv')
```

```python
# Loop through the years from 2005 to 2011
for year in range(2005, 2012):
    # Create a horizontal bar graph of all suicides by state for the current year
    plt.figure(figsize=(12,8))
    df_sorted = df.sort_values(by='state', ascending=False)
    plt.barh(df_sorted['state'], df_sorted[f'{year}_all_suicides'], align='center')
    plt.xlim(0, 4000)
    plt.xlabel(f'{year}_all_suicides')
    plt.ylabel('State')
    plt.title(f'Suicides by State in {year}')

    # Set the y-tick labels to the 'state' column of the sorted DataFrame
    plt.yticks(df_sorted['state'])

    plt.show()
```

```python
# Loop through the years from 2005 to 2011 for vet
for year in range(2005, 2012):
    # Create a horizontal bar graph of vet suicides by state for the current year
    plt.figure(figsize=(12,8))
    plt.barh(df_sorted['state'], df_sorted[f'{year}_vet_suicides'], align='center')
    plt.xlim(0, 800)
    plt.xlabel(f'{year}_vet_suicides')
    plt.ylabel('State')
    plt.title(f'Vet Suicides by State in {year}')

    # Set the y-tick labels to the 'state' column of the sorted DataFrame
    plt.yticks(df_sorted['state'])

    plt.show()
    
    # Create a scatter plot of all suicides vs. veteran suicides for the current year
    plt.figure(figsize=(12,8))
    plt.scatter(df[f'{year}_all_suicides'], df[f'{year}_vet_suicides'])
    plt.xlabel(f'{year}_all_suicides')
    plt.ylabel(f'{year}_vet_suicides')
    plt.title(f'All Suicides vs. Veteran Suicides in {year}')

    # Add a regression line to the scatter plot
    slope, intercept, r_value, p_value, std_err = linregress(df[f'{year}_all_suicides'], df[f'{year}_vet_suicides'])
    x = np.linspace(0, max(df[f'{year}_all_suicides']), 100)
    y = slope * x + intercept
    plt.plot(x, y, color='r')

    plt.show()
```

```python
# Load the data from the CSV file
data = pd.read_csv('combined_cleaned.csv')

# Select the columns of interest
all_columns = ['2005_all_suicides', '2006_all_suicides', '2007_all_suicides',
               '2008_all_suicides', '2009_all_suicides', '2010_all_suicides', '2011_all_suicides']
vet_columns = ['2005_vet_suicides', '2006_vet_suicides', '2007_vet_suicides',
               '2008_vet_suicides', '2009_vet_suicides', '2010_vet_suicides', '2011_vet_suicides']

# Loop over each year and create the histogram
for year in range(len(all_columns)):
    column_all = all_columns[year]
    column_vet = vet_columns[year]
    plt.hist(data[column_all], bins=10, alpha=0.5, label='All')
    plt.hist(data[column_vet], bins=10, alpha=0.5, label='Veterans')
    plt.title(f'{column_all} vs. {column_vet}')
    plt.xlabel('Number of Suicides')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()
```
